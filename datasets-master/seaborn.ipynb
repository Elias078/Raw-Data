```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
a = sns.load_dataset("flights")
a.head(200)
```

```python
sns.relplot(x="passengers", y="month", data=a)
```

```python
sns.relplot(x="passengers", y="month",hue = "year", data=a)
```

```python
b = sns.load_dataset("tips")
b.head()
```

```python
sns.relplot(x='time', y='tip', data = b, kind='line', hue="sex")
```

```python
sns.catplot(x='day', y='total_bill', data=b, hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=b, kind="boxen")
```
